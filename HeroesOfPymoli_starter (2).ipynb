## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = 0, 10, 14, 19, 24, 29, 34, 39, 40
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40"]
```

```python
pd.cut(purchase_data["Age"], bins, labels=bin_names, include_lowest=True)
```

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=bin_names)
purchase_data.head()
```

```python
age_group = purchase_data.groupby("Age Group")

print(age_group["Age Group"].count())
```

```python
summary_age = pd.DataFrame(age_group.count())
summary_age
```

```python
summary_age = summary_age[["Purchase ID", "Age"]]
summary_age
```

```python
summary_age["Purchase ID"] = ((summary_age["Purchase ID"]/total_players)*100)
summary_age
```

```python
summary_age["Purchase ID"] = summary_age["Purchase ID"].astype(float).map("{:,.2f}%".format)
summary_age
```

```python
summary_age = summary_age.rename(columns={"Purchase ID":"Percentage of Players", "Age":"Total Count"})
summary_age
```
